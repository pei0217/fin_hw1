```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# 設定目標網址（Yahoo 奇摩台積電頁面）
stock_id = '2317'

url = f'https://tw.stock.yahoo.com/quote/{stock_id}'

# 發送請求
response = requests.get(url)

# 解析網頁內容
soup = BeautifulSoup(response.text, 'html.parser')

date = soup.find('time')
all_data = soup.find('ul', {'class': 'D(f) Fld(c) Flw(w) H(192px) Mx(-16px)'}).find_all('li')  # 這裡需要根據實際網頁結構修改

fields=[]
data=[]

for d in all_data:
  name = d.find_all('span')
  for num,name in enumerate(name):
    if name.text == '': # 表格內容空就跳過
      continue

    if num == 0:
     fields.append(name.text)
    else:
     data.append(name.text)
df = pd.DataFrame([data], columns=fields)
df.insert(0,'日期',date['datatime'])
df.insert(1,'股號',stock_id)


df
```
